# Q2. Cosine similarity
#### The embedder returns normalized vectors, so the dot product between two of them is their cosine similarity.

#### Take the page 02-vector-search/lessons/07-sqlitesearch-vector.md, embed its content, and compute the cosine similarity with the query vector from Q1. What do you get?

###### 0.07
###### 0.37
###### 0.68
###### 0.92


```python
import subprocess

code = """
from gitsource import GithubRepositoryDataReader
from embedder import Embedder
import numpy as np

# 1. Load the data using the pinned commit
reader = GithubRepositoryDataReader(
    repo_owner="DataTalksClub",
    repo_name="llm-zoomcamp",
    commit_id="8c1834d",
    allowed_extensions={"md"},
    filename_filter=lambda path: "/lessons/" in path,
)
documents = [file.parse() for file in reader.read()]

# 2. Find the target document
target_file = "02-vector-search/lessons/07-sqlitesearch-vector.md"
doc = next(d for d in documents if d['filename'] == target_file)

# 3. Initialize embedder and encode both text strings
embedder = Embedder()
v_query = embedder.encode("How does approximate nearest neighbor search work?")
v_doc = embedder.encode(doc['content'])

# 4. Calculate Cosine Similarity (Dot Product because they are normalized)
similarity = np.dot(v_query, v_doc)
print(f"SIM_RESULT: {similarity:.4f}")
"""

cmd = ["/Users/nikhilchoudhary/llm-zoomcamp-2026/.venv/bin/python", "-c", code]
result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
```
